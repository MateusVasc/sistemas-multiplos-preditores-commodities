```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.data.loaders import CommodityLoader
```

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
def plot_all_data(df):
    for u in df['unique_id'].unique():
        plt.figure(figsize=(12,6))
        plt.plot(df[df['unique_id'] == u]['ds'], df[df['unique_id'] == u]['y'], label=u)
        plt.title(f'Série Temporal - {u}')
        plt.xlabel('Date')
        plt.ylabel('Price (y)')
        plt.legend()
        plt.grid(True)
        plt.show()
```

### Daily

```python
df_brl = CommodityLoader.load_all_commodities(currency='BRL', preprocessing=True, monthly_aggregation=None, limit_date=None)
```

```python
df_brl.head(15)
```

```python
plot_all_data(df_brl)
```
