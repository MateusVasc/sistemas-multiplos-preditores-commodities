```python
def extract_lag_windows(series, windows_size):
    windows, targets = [], []
    for i in range(windows_size, len(series)):
        windows.append(series[i-windows_size:i])
        targets.append(series[i])
    return np.array(windows), np.array(targets)
```

```python
acucar_series = df_acucar['y']
print(type(acucar_series))
print(len(acucar_series))
```

```python
win, tar = extract_lag_windows(acucar_series, 12)
```

```python
df_win = pd.DataFrame(
    win,
    columns=[f"lag_{i}" for i in range(1, win.shape[1] + 1)]
)
df_win["target"] = tar
n_lags = win.shape[1]
df_win.columns = [f"lag_t-{n_lags - i}" for i in range(n_lags)] + ["target"]

df_win
```
