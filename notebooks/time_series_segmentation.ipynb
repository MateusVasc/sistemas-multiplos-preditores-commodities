### Separando os dataframes pelo tipo do índice (PREÇO ou QUANTUM)

```python
def get_series_by_tipo_indice(df):
    df_preco = df[df['TIPO_INDICE'] == 'PRECO']
    df_quantum = df[df['TIPO_INDICE'] == 'QUANTUM']

    return df_preco, df_quantum
```

```python
df_preco, df_quantum = get_series_by_tipo_indice(df)
```

### Separando os dataframes pelo tipo (CORRENTE, EXP ou IMP)

```python
def get_series_by_tipo(df):
    df_corr = df[df['TIPO'] == 'CORRENTE']
    df_exp = df[df['TIPO'] == 'EXP']
    df_imp = df[df['TIPO'] == 'IMP']

    return df_corr, df_exp, df_imp
```

### Separando os dataframes pelo tipo de commodities (Combustíveis, Matérias-Primas Agrícolas, Minerais, Minérios E Metais, Todos Alimentos, Total Commodities)

```python
def get_series_by_tipo(df):
    df_comb = df[df['NO_COMMODITIES'] == 'Combustíveis']
    df_mpa = df[df['NO_COMMODITIES'] == 'Matérias-Primas Agrícolas']
    df_mmm = df[df['NO_COMMODITIES'] == 'Minerais, Minérios E Metais']
    df_ta = df[df['NO_COMMODITIES'] == 'Todos Alimentos']
    df_tc = df[df['NO_COMMODITIES'] == 'Total Commodities']

    return df_comb, df_mpa, df_mmm, df_ta, df_tc
```
